```python
# Import Libraries
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
## Remember Income column has Nan Values
```

```python
# Reading Data
data = pd.read_csv('marketing_campaign.csv',delimiter='\t')
data
```

```python
# Data Headers And Tails
print("Data Headers")
print(data.head(),'\n')
print("Data Tails")
print(data.tail(),'\n')
```

```python
# Data Types
print(data.dtypes)
```

```python
# Information And Describtion About Data
data = data.drop(columns=["ID"]) # Dropping ID
print("Data Describtion")
print(data.describe())
```

```python
# Information About Data
print("Data Information")
print(data.info())
```

```python
# Checking Missing Value
print("Checking Missing Data Values")
print(data.isnull().sum(),'\n')
print("Check Duplicated ")
print(data.duplicated().sum())
# Only if existed duplicated but now no found 0
# data = data.drop_duplicates()
```

```python
# Distribution (Univariate Analysis)
px.histogram(data, x='Year_Birth', nbins=20, title='Age Distribution').show()
px.histogram(data, x='Income', nbins=30, title='Income Distribution').show()
data['Total_Spending'] = data[['MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']].sum(axis=1)
px.histogram(data, x='Total_Spending', nbins=30, title='Total Spending Distribution').show()
```

```python
# Categorical Analysis
px.bar(data, x='Education', title='Education Levels').show()
px.bar(data, x='Marital_Status', title='Marital Status Count').show()
```

```python
# Web & Store Behavior
px.scatter(data, x='NumWebVisitsMonth', y='NumWebPurchases', color='Marital_Status', title='Web Visits vs Web Purchases').show()
px.scatter(data, x='NumStorePurchases', y='Total_Spending', color='Marital_Status', title='Store Purchases vs Total Spending').show()
```
